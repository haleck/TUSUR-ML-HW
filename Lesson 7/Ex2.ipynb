```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Olympics 2018.xlsx', sheet_name='Olympics', usecols='A:Q')
df.head()
```

```python
russian_gold_medalists = df.query("Country == 'Russia' and Medal == 'GOLD'")
russian_gold_medalists.head()
```

```python
import matplotlib.pyplot as plt
```

```python
ax = russian_gold_medalists.plot.scatter(x='Age', y='Sport', figsize=(16,8))
ax.figure.savefig('Figure1.pdf')
```

```python
ax = df.groupby('Year').Age.mean().dropna().plot()
ax.figure.savefig('Figure2.pdf')
```

```python
summer_season = df.query("Season == 'Summer'")
summer_season
```

```python
summer_season.drop(columns=['City'], inplace=True)
summer_season
```

```python
summer_season['Year of birth'] = summer_season['Year'] - summer_season['Age']
summer_season
```

```python
summer_season.to_csv('Olympics-summer-season.csv')
```
